```python
from kubernetes import client,config

config.load_kube_config(config_file="./config")
api_instance = client.CoreV1Api()
list_deployment_ai = client.AppsV1Api()
```

```python
deployments = list_deployment_ai.list_namespaced_deployment(namespace='service-predict').items
for deployment in deployments:
    print(deployment.metadata.name)
    print(deployment.spec.replicas)
    print(deployment.spec.template)
    print(deployment.status.available_replicas)
    print('---------------------------------------------------------------')
```

```python
import ast

items = api_instance.list_namespaced_pod(namespace='service-predict').items
print(f"The number of pods in service-predict namespace is: {len(items)}")
for item in items:
    print(item.metadata.generate_name)
    print(item.metadata.labels)
#     print(item.spec)
    
    print(ast.literal_eval(str(item.spec)))
    break
```

```python
services = api_instance.list_namespaced_service(namespace="service-predict").items
```

```python
print(f"The number of services in service-predict namespace is: {len(services)}")
for service in services:
    print(service.metadata.name)
    print(service.spec)
    print("----------------------------------------------------------------------")
```
